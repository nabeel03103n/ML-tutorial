```python
df4 = df3.copy()
df4.head()
```

```python
# df4.ST_Slope.replace({
#     'Down':1,'Flat':2,'Up':3
# },inplace=True)

df5 = pd.get_dummies(df4,drop_first=True).astype(int)
df5.head()
```

```python
X = df5.drop('HeartDisease',axis='columns')
y = df5.HeartDisease
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rr = RandomForestClassifier()
rr.fit(X_train,y_train)
rr.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X)
X_pca
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_pca,y,test_size=0.2)
```

```python
rr.fit(X_train,y_train)
rr.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(RandomForestClassifier(),X_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
cross_val_score(KNeighborsClassifier(),X_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression
cross_val_score(LogisticRegression(max_iter=400),X_train,y_train)
```
