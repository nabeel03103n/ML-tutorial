# 2.1 Gradient Descent and Cost Function

Imagine you're at the top of a hill
You’re standing on a hill, and your goal is to get to the bottom where there’s a big bowl of candy (yay!). The bottom of the hill is the sweet spot, and that’s what we call the lowest cost.

What’s the cost function?
The hill is shaped like a slide, but bumpy. The cost function is like a map that tells us how high we are on the hill. Higher up = more "cost" (farther from candy). Lower down = less "cost" (closer to candy).

What’s gradient descent?
Now, to get to the candy, you can’t jump or fly. You have to take small steps down the hill. Every time you take a step, you look around to find which way is the steepest downhill (the fastest way to get closer to the candy). That’s gradient descent: taking steps in the direction that reduces your cost (or height) the most.

Why small steps?
If you take big steps, you might tumble too far and miss the candy. Small steps help you move carefully and reach the sweet spot exactly.

Putting it together:
Cost function = The map of the hill showing how far you are from the candy.
Gradient descent = Taking careful steps downhill to reach the candy.

# Learning Rate

    m = m - learning rate * ∂/∂m
    b = b - learning rate * ∂/∂b


What is Learning Rate in Machine Learning?
The learning rate is a crucial hyperparameter in machine learning that controls how much a model adjusts its weights during training in response to the error (or loss) it experiences. It's often denoted as α or η.



In the upper image you are on b1 and you want to come to b2 so you will subtract the learning rate * ∂/∂b to get the b2 <br>

    b_new = b_old - learning rate * ∂/∂b

```python
# Now lets start making a function
import numpy as np

def gradient_descent(x,y):
    m_curr = b_curr = 0 # Starting value of m and b
    iteration = 10000
    n = len(x) # Assuming that the lenth of x and y are same
    learning_rate = 0.08 # changing this, will affect the cost

    for i in range(iteration):
        y_predicted = m_curr * x + b_curr # y = mx+b

        # Derivative part
        md = -(2/n)*sum(x*(y-y_predicted)) # Formula ∂/∂m = 2/n Σⁿ -xi(yi - yp)
        bd = -(2/n)*sum(y-y_predicted) # Formula ∂/∂m = 2/n Σⁿ -(yi - yp)
    
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)]) # MSE (Cost function) formula: mse = 1/n Σⁿ(yi - yp)² 

        # Now update the m using the formula m =  m - learning rate * ∂/∂m also b
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print(f"m {m_curr}, b {b_curr}, iteration {i}, cost {cost}") # The cost is decreasing
        

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08
    costs = []
    iterations_list = []

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        md = -(2 / n) * sum(x * (y - y_predicted))
        bd = -(2 / n) * sum(y - y_predicted)
        cost = (1 / n) * sum([val**2 for val in (y - y_predicted)])

        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        costs.append(cost)
        iterations_list.append(i)

    # Plot the cost reduction with logarithmic scale
    plt.figure(figsize=(10, 6))
    plt.plot(iterations_list, costs, label="Cost (MSE)", color="blue")
    plt.yscale("log")  # Logarithmic scale for better visualization
    plt.xlabel("Iterations")
    plt.ylabel("Cost (Log Scale)")
    plt.title("Cost Reduction Over Iterations (Log Scale)")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Zoom into the first 1000 iterations
    plt.figure(figsize=(10, 6))
    plt.plot(iterations_list[:1000], costs[:1000], label="Cost (MSE)", color="red")
    plt.xlabel("Iterations (First 1000)")
    plt.ylabel("Cost")
    plt.title("Cost Reduction in First 1000 Iterations")
    plt.legend()
    plt.grid(True)
    plt.show()

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```
