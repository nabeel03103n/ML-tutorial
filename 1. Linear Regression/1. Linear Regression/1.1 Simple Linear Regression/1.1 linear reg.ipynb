## Simple Linar Reg

we will do machine learning in this notebook using simple linear reg with only one features to predict the label

```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# this line is needed to display the plot in the notebook

plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,marker='+',color='red')
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression() # create a linear regression object
reg.fit(df[['area']],df.price) # train the model
```

```python
reg.predict([[3300]]) # predict the price of a house with 3300 sqr ft area
```

Its predicting the price using the formual y = mx+b

x is the area y is the price and m is a coeff

price = m * area + b

here price is the dependent variable and area is a independent variable


```python
reg.coef_ # get the value of the coefficient
```

```python
reg.intercept_ # get the value of the intercept (b)
```

```python
#lets calculate the price for the following areas
# y = mx+b
135.78767123*3300+180616.43835616432
```

```python
d = pd.read_csv('area.csv')
d.head()
```

```python
p = reg.predict(d)
```

```python
d['price'] = p
```

```python
d.to_csv('prediction.csv',index=False)
```

```python
# now lets plot the prediction line
plt.xlabel('area')
plt.ylabel('predicted price')
plt.scatter(df.area,df.price,marker='+',color='red')
plt.plot(df.area,reg.predict(df[['area']]))
```
